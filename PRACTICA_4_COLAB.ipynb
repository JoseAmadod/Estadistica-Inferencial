```python
#PUNTO 1

import numpy as np
from scipy import stats

# Given data
sample_mean = 170  # cm
population_std_dev = 10  # cm
sample_size = 50
confidence_level = 0.95

# Calculate the critical value (Z-score) for the given confidence level
alpha = 1 - confidence_level
z_critical = stats.norm.ppf(1 - alpha/2)

# Calculate the margin of error
margin_of_error = z_critical * (population_std_dev / np.sqrt(sample_size))

# Calculate the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

# Print the results
print(f"Sample Mean: {sample_mean} cm")
print(f"Population Standard Deviation: {population_std_dev} cm")
print(f"Sample Size: {sample_size}")
print(f"Confidence Level: {confidence_level}")
print(f"Z-critical value: {z_critical:.4f}")
print(f"Margin of Error: {margin_of_error:.4f} cm")
print(f"95% Confidence Interval: ({confidence_interval_lower:.4f}, {confidence_interval_upper:.4f}) cm")
```

# Task
Perform a hypothesis test to determine if the average weight of cereal packages is less than 500g, given a sample mean of 495g, a population standard deviation of 15g, a sample size of 40, and a significance level of 0.05.

## Define hypotheses

### Subtask:
State the null and alternative hypotheses based on the problem description.


**Reasoning**:
State the null and alternative hypotheses based on the problem description.



```python
#PUNTO 21
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Define the null hypothesis (H0)
h0 = "H0: The average weight of cereal packages is equal to 500g (μ = 500g)"

# Define the alternative hypothesis (H1)
h1 = "H1: The average weight of cereal packages is less than 500g (μ < 500g)"

# Print the hypotheses
print(h0)
print(h1)

# Given data
sample_mean = 495  # g
population_mean = 500  # g
population_std_dev = 15  # g
sample_size = 40
significance_level = 0.05

# Calculate the standard error of the mean
standard_error = population_std_dev / np.sqrt(sample_size)

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / standard_error

# Calculate the critical value (Z-score) for a one-tailed test to the left
alpha = significance_level
z_critical = stats.norm.ppf(alpha)

# Print the calculated Z-statistic and critical Z-value
print(f"\nSample Mean: {sample_mean} g")
print(f"Population Mean: {population_mean} g")
print(f"Population Standard Deviation: {population_std_dev} g")
print(f"Sample Size: {sample_size}")
print(f"Standard Error: {standard_error:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"Significance Level (alpha): {alpha}")
print(f"Critical Z-value (one-tailed left): {z_critical:.4f}")

# Define the rejection region
print(f"Rejection Region: Z < {z_critical:.4f}")

# Compare the Z-statistic to the critical Z-value
if z_statistic < z_critical:
    conclusion = "Reject the null hypothesis"
    reason = f"The calculated Z-statistic ({z_statistic:.4f}) is less than the critical Z-value ({z_critical:.4f}), falling in the rejection region."
else:
    conclusion = "Fail to reject the null hypothesis"
    reason = f"The calculated Z-statistic ({z_statistic:.4f}) is not less than the critical Z-value ({z_critical:.4f}), and does not fall in the rejection region."

# Print the conclusion and reason
print("\nConclusion:")
print(conclusion)
print(reason)

# Plotting the results
plt.figure(figsize=(10, 6))
x = np.linspace(-4, 4, 1000)
plt.plot(x, stats.norm.pdf(x, 0, 1), label='Standard Normal Distribution')
plt.axvline(z_critical, color='red', linestyle='--', label=f'Critical Z-value ({z_critical:.4f})')
plt.axvline(z_statistic, color='green', linestyle='-', label=f'Z-statistic ({z_statistic:.4f})')

# Shade the rejection region
x_reject = np.linspace(-4, z_critical, 100)
plt.fill_between(x_reject, 0, stats.norm.pdf(x_reject, 0, 1), color='red', alpha=0.2, label='Rejection Region')

plt.title('Hypothesis Test for Cereal Package Weight (PUNTO 21)')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#PUNTO 31

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Define the null hypothesis (H0)
h0 = "H0: The population variance of call duration is 2.5 minutes² (σ² = 2.5)"

# Define the alternative hypothesis (H1)
h1 = "H1: The population variance of call duration differs from 2.5 minutes² (σ² ≠ 2.5)"

# Print the hypotheses
print(h0)
print(h1)

# Given data
population_variance_h0 = 2.5  # σ² under the null hypothesis
sample_variance = 4  # s²
sample_size = 16
significance_level = 0.05

# Calculate the chi-squared statistic
# Formula: χ² = ((n - 1) * s²) / σ²
chi_squared_statistic = ((sample_size - 1) * sample_variance) / population_variance_h0

# Determine degrees of freedom
degrees_of_freedom = sample_size - 1

# Determine critical values for a two-tailed test
alpha = significance_level
chi_squared_critical_lower = stats.chi2.ppf(alpha/2, degrees_of_freedom)
chi_squared_critical_upper = stats.chi2.ppf(1 - alpha/2, degrees_of_freedom)

# Print the calculated chi-squared statistic, degrees of freedom, and critical values
print(f"\nPopulation Variance (H0): {population_variance_h0} minutes²")
print(f"Sample Variance: {sample_variance} minutes²")
print(f"Sample Size: {sample_size}")
print(f"Chi-squared statistic: {chi_squared_statistic:.4f}")
print(f"Degrees of Freedom: {degrees_of_freedom}")
print(f"Significance Level (alpha): {alpha}")
print(f"Critical Chi-squared values (two-tailed): {chi_squared_critical_lower:.4f} and {chi_squared_critical_upper:.4f}")

# Define the rejection region
print(f"Rejection Region: χ² < {chi_squared_critical_lower:.4f} or χ² > {chi_squared_critical_upper:.4f}")

# Conclude if H0 is rejected
if chi_squared_statistic < chi_squared_critical_lower or chi_squared_statistic > chi_squared_critical_upper:
    conclusion = "Reject the null hypothesis"
    reason = f"The calculated Chi-squared statistic ({chi_squared_statistic:.4f}) falls in the rejection region (χ² < {chi_squared_critical_lower:.4f} or χ² > {chi_squared_critical_upper:.4f})."
else:
    conclusion = "Fail to reject the null hypothesis"
    reason = f"The calculated Chi-squared statistic ({chi_squared_statistic:.4f}) does not fall in the rejection region (χ² < {chi_squared_critical_lower:.4f} or χ² > {chi_squared_critical_upper:.4f})."

# Print the conclusion and reason
print("\nConclusion:")
print(conclusion)
print(reason)

# Plotting the results (Chi-squared distribution)
plt.figure(figsize=(10, 6))
x = np.linspace(0, stats.chi2.ppf(0.999, degrees_of_freedom), 1000) # Adjust x-limit for better visualization
plt.plot(x, stats.chi2.pdf(x, degrees_of_freedom), label=f'Chi-squared Distribution (df={degrees_of_freedom})')
plt.axvline(chi_squared_critical_lower, color='red', linestyle='--', label=f'Critical χ² Lower ({chi_squared_critical_lower:.4f})')
plt.axvline(chi_squared_critical_upper, color='red', linestyle='--', label=f'Critical χ² Upper ({chi_squared_critical_upper:.4f})')
plt.axvline(chi_squared_statistic, color='green', linestyle='-', label=f'χ² statistic ({chi_squared_statistic:.4f})')

# Shade the rejection regions
x_reject_lower = np.linspace(0, chi_squared_critical_lower, 100)
x_reject_upper = np.linspace(chi_squared_critical_upper, stats.chi2.ppf(0.999, degrees_of_freedom), 100)
plt.fill_between(x_reject_lower, 0, stats.chi2.pdf(x_reject_lower, degrees_of_freedom), color='red', alpha=0.2, label='Rejection Region')
plt.fill_between(x_reject_upper, 0, stats.chi2.pdf(x_reject_upper, degrees_of_freedom), color='red', alpha=0.2)


plt.title('Hypothesis Test for Variance of Call Duration (PUNTO 31)')
plt.xlabel('Chi-squared statistic')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```
