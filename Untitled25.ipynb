```python
# 3. Eliminar filas con datos faltantes y verificar limpieza

# Eliminar filas con datos faltantes
df_filtrado = df.dropna()

# Mostrar el número de filas después de la limpieza
print(f"\nNúmero de filas después de eliminar datos faltantes: {df_filtrado.shape[0]}")

# Identificar y mostrar el número de datos faltantes por columna
faltantes = df_filtrado.isnull().sum()
print("\nDatos faltantes por columna después de la limpieza:")
print(faltantes)

# Visualizar el mapa de calor de datos faltantes (debería estar vacío si la limpieza fue exitosa)
plt.figure(figsize=(10, 2))
sns.heatmap(df_filtrado.isnull(), cbar=False, cmap="viridis")
plt.title("Mapa de datos faltantes después de la limpieza")
plt.tight_layout()
# plt.savefig("mapa_datos_faltantes.png") # Quitar comentario si se quiere guardar la figura
plt.show()


# Identificar variables categóricas y numéricas
categorical_cols = df_filtrado[df_columnas].select_dtypes(include=['object']).columns
numeric_cols = df_filtrado[df_columnas].select_dtypes(include=np.number).columns

# Generar gráficos para variables categóricas
for col in categorical_cols:
    print(f"\nGráficos para la columna: {col}")

    # Diagrama de barras
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df_filtrado[df_columnas], y=col, order=df_filtrado[df_columnas][col].value_counts().index, palette='viridis')
    plt.title(f'Distribución de {col}')
    plt.xlabel('Frecuencia')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

    # Diagrama de torta (solo si hay un número razonable de categorías)
    if df_filtrado[df_columnas][col].nunique() <= 15: # Limitar a 15 categorías para evitar gráficos de torta ilegibles
        plt.figure(figsize=(8, 8))
        df_filtrado[df_columnas][col].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='viridis')
        plt.title(f'Distribución porcentual de {col}')
        plt.ylabel('') # Eliminar etiqueta del eje y para el gráfico de torta
        plt.show()
    else:
        print(f"No se generó diagrama de torta para {col} debido a la gran cantidad de categorías.")


# Generar gráficos para variables numéricas
for col in numeric_cols:
    print(f"\nGráficos para la columna: {col}")

    # Histograma
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_filtrado[df_columnas], x=col, kde=True, palette='viridis')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()

    # Diagrama de caja
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df_filtrado[df_columnas], x=col, palette='viridis')
    plt.title(f'Diagrama de Caja de {col}')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
# Seleccionar solo las columnas numéricas para el cálculo del IQR
numeric_cols_for_outliers = df_filtrado[numeric_cols]

# Calcular el Q1 (primer cuartil) y Q3 (tercer cuartil)
Q1 = numeric_cols_for_outliers.quantile(0.25)
Q3 = numeric_cols_for_outliers.quantile(0.75)

# Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los límites para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identificar outliers
outliers = ((numeric_cols_for_outliers < lower_bound) | (numeric_cols_for_outliers > upper_bound)).any(axis=1)

# Filtrar el DataFrame para eliminar filas con outliers
df_cleaned = df_filtrado[~outliers]

print(f"Número de filas antes de eliminar outliers: {df_filtrado.shape[0]}")
print(f"Número de filas después de eliminar outliers: {df_cleaned.shape[0]}")

# Opcional: Mostrar las filas que fueron identificadas como outliers
# print("\nFilas identificadas como outliers:")
# display(df_filtrado[outliers])
```

```python
# Generar histogramas para variables numéricas después de eliminar outliers
numeric_cols = df_cleaned.select_dtypes(include=np.number).columns

for col in numeric_cols:
    print(f"\nHistograma para la columna: {col} (sin outliers)")

    plt.figure(figsize=(10, 6))
    sns.histplot(data=df_cleaned, x=col, kde=True)
    plt.title(f'Histograma de {col} (sin outliers)')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.tight_layout()
    plt.show()
```

```python
from scipy import stats
import statsmodels.api as sm

# Identificar variables numéricas continuas (excluyendo 'MODELO' y 'AÑO_REGISTRO' si se consideran discretas)
# Basándonos en la inspección inicial, consideraremos 'PESO', 'POTENCIA', 'CILINDRAJE', 'CAPACIDAD_CARGA', 'CAPACIDAD_PASAJEROS' como potencialmente continuas.
continuous_numeric_cols = ['PESO', 'POTENCIA', 'CILINDRAJE', 'CAPACIDAD_CARGA', 'CAPACIDAD_PASAJEROS']

# Asegurarse de que las columnas existan en el DataFrame limpio
continuous_numeric_cols = [col for col in continuous_numeric_cols if col in df_cleaned.columns]

print("Realizando análisis de normalidad para las siguientes columnas:")
print(continuous_numeric_cols)

for col in continuous_numeric_cols:
    print(f"\nAnálisis de normalidad para la columna: {col}")

    # Eliminar posibles valores NaN que puedan haber quedado (aunque después de dropna y IQR, deberían ser pocos o ninguno)
    data = df_cleaned[col].dropna()

    if len(data) > 3: # Algunos tests requieren al menos 4 puntos de datos
        # 1) Test de Shapiro-Wilk
        if len(data) >= 3 and len(data) <= 5000: # Shapiro-Wilk es más adecuado para muestras pequeñas (n <= 5000)
            shapiro_test = stats.shapiro(data)
            print(f"  Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")
            if shapiro_test.pvalue < 0.05:
                print("    --> La distribución NO parece normal (rechazar H0)")
            else:
                print("    --> La distribución parece normal (no rechazar H0)")
        else:
            print("  Shapiro-Wilk Test: No aplicable o muestra demasiado grande para este test (n > 5000).")

        # 2) Test de Kolmogorov-Smirnov (Lilliefors para distribución subyacente desconocida)
        if len(data) > 0:
          try:
              kstest_test = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
              print(f"  Kolmogorov-Smirnov Test: Statistic={kstest_test.statistic:.4f}, p-value={kstest_test.pvalue:.4f}")
              if kstest_test.pvalue < 0.05:
                  print("    --> La distribución NO parece normal (rechazar H0)")
              else:
                  print("    --> La distribución parece normal (no rechazar H0)")
          except ValueError as e:
              print(f"  Kolmogorov-Smirnov Test: Error - {e}")
        else:
            print("  Kolmogorov-Smirnov Test: No aplicable (no hay datos).")


        # 3) Test de Anderson-Darling
        if len(data) > 7: # Anderson-Darling requiere al menos 8 puntos de datos
            try:
                anderson_test = stats.anderson(data, dist='norm')
                print(f"  Anderson-Darling Test:")
                print(f"    Statistic: {anderson_test.statistic:.4f}")
                for i in range(len(anderson_test.critical_values)):
                    sig_level = anderson_test.significance_level[i]
                    critical_value = anderson_test.critical_values[i]
                    print(f"    Critical Value at {sig_level}% significance: {critical_value:.4f}")
                if anderson_test.statistic > anderson_test.critical_values[anderson_test.significance_level.tolist().index(5.0)]: # Comparar con el valor crítico al 5%
                     print("    --> La distribución NO parece normal (estadístico > valor crítico)")
                else:
                    print("    --> La distribución parece normal (estadístico <= valor crítico)")
            except ValueError as e:
                print(f"  Anderson-Darling Test: Error - {e}")
        else:
             print("  Anderson-Darling Test: No aplicable (menos de 8 datos).")


        # 4) Gráfico QQ
        print("  Generando gráfico QQ...")
        plt.figure(figsize=(8, 6))
        sm.qqplot(data, line='s') # 's' for standardized line
        plt.title(f'QQ Plot de {col}')
        plt.show()

        # 5) Test de Jarque-Bera
        if len(data) > 2: # Jarque-Bera requiere al menos 3 datos
            try:
                jb_test = stats.jarque_bera(data)
                print(f"  Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")
                if jb_test.pvalue < 0.05:
                    print("    --> La distribución NO parece normal (rechazar H0)")
                else:
                    print("    --> La distribución parece normal (no rechazar H0)")
            except ValueError as e:
                print(f"  Jarque-Bera Test: Error - {e}")
        else:
            print("  Jarque-Bera Test: No aplicable (menos de 3 datos).")

    else:
        print(f"  No hay suficientes datos en la columna {col} para realizar pruebas de normalidad.")
```
